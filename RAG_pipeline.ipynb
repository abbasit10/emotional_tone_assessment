# LLAMA RAG

```python
# Loading pickle file containing all documents
with open("/content/drive/MyDrive/all_docs.pkl", "rb") as f:
    all_docs = pickle.load(f)

print("Total documents loaded:", len(all_docs))
```

```python
# Embedding model
Settings.embed_model = HuggingFaceEmbedding(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
# Chunking
CHUNK_SIZE = 400
CHUNK_OVERLAP = 70

node_parser = SentenceSplitter(
    chunk_size=CHUNK_SIZE,
    chunk_overlap=CHUNK_OVERLAP
)
```

```python
# Converting documents to chunks
nodes = node_parser.get_nodes_from_documents(all_docs)

print("Total chunks created:", len(nodes))
```

```python
# LLM(GROQ) SetuP
GROQ_API_KEY = userdata.get("GROQ_API_KEY")
os.environ["GROQ_API_KEY"] = GROQ_API_KEY

llm = Groq(
    model="llama-3.1-8b-instant",
    groq_api_key=GROQ_API_KEY,
    temperature=0
)

Settings.llm = llm
```

```python
# Building vector Index
index = VectorStoreIndex(nodes)
```

```python
# Creating retriever
retriever = index.as_retriever(similarity_top_k=2)
```

```python
# Retrieving context

def retrieve_context(question):

    nodes = retriever.retrieve(question)
    context = "\n".join([n.node.text for n in nodes])

    return context
```

```python
# LLM generation

def generate_answer(question, context):

    prompt = f"""
Use the following context to answer the question.

Context:
{context}

Question:
{question}

Answer:
"""

    response = llm.complete(prompt)

    time.sleep(6)

    return str(response)
```

```python
# RAG Pipeline
def run_rag(df, question_col):

    df_out = df.copy()

    # Retrieve contexts
    print("Retrieving contexts...")
    contexts = [retrieve_context(q) for q in tqdm(df_out[question_col].astype(str))]

    # Generate answers in parallel
    print("Generating answers with LLM...")

    def process(i):
        q = df_out.iloc[i][question_col]
        c = contexts[i]
        return generate_answer(q, c)

    with ThreadPoolExecutor(max_workers=2) as executor:
        answers = []
        for i in tqdm(range(len(df_out))):
            answers.append(process(i))

    df_out["rag_answer"] = answers

    return df_out

```

```python
# Randomly select 500 rows
df1_sample = df1_final.sample(n=500, random_state=42)

# Run RAG on the sampled dataframe
df1_final_rag = run_rag(df1_sample, "text")

# View results
df1_final_rag.head()
```

# QWEN RAG

```python
# LLM Setup
model_name = "Qwen/Qwen2.5-3B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)

llm = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto" if device == "cuda" else None
)

llm = llm.to(device)
```

```python
# Building vector Index
index = VectorStoreIndex(nodes)
```

```python
# Creating retriever
retriever = index.as_retriever(similarity_top_k=2)
```

```python
# Retrieving context

def retrieve_context(question):

    nodes = retriever.retrieve(question)
    context = "\n".join([n.node.text for n in nodes])

    return context[:1500]
```

```python
# Function to generate a complete answer using a LLM
def generate_answer(question, context):

    prompt = f"""
You are helpful assistant.
Use the following context to answer the question.

Rules:
- - Write a clear and complete answer.
- Do NOT leave the ending sentence incomplete.

Context:
{context}

Question:
{question}

Answer:
Write a complete answer.
"""
    # Tokenizing the prompt
    inputs = tokenizer(prompt, return_tensors="pt").to(device)

    # Generating response
    outputs = llm.generate(
    **inputs,
    max_new_tokens=300,
    temperature=0.0,
    do_sample=False,
    pad_token_id=tokenizer.eos_token_id
    )

    response = tokenizer.decode(
    outputs[0][inputs["input_ids"].shape[-1]:],
    skip_special_tokens=True
)

    response = response.strip()

    # Keep only complete sentences
    import re
    sentences = re.split(r'(?<=[.!?])\s+', response)
    response = " ".join(sentences[:-1]) if not response.endswith(('.', '!', '?')) else response

    return response
```

```python
# RAG Pipeline
def run_rag(df, question_col):

    df_out = df.copy()

    # Retrieve contexts
    print("Retrieving contexts...")
    contexts = [retrieve_context(q) for q in tqdm(df_out[question_col].astype(str))]

    # Generate answers in parallel
    print("Generating answers with LLM...")

    def process(i):
        q = df_out.iloc[i][question_col]
        c = contexts[i]
        return generate_answer(q, c)

    with ThreadPoolExecutor(max_workers=2) as executor:
        answers = []
        for i in tqdm(range(len(df_out))):
            answers.append(process(i))

    df_out["rag_answer"] = answers

    return df_out

```
